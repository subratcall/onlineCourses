```python
from models.AE import Autoencoder
import os
```

```python
from utils.loaders import load_mnist
```

```python
(x_train, y_train), (x_test, y_test) = load_mnist()
```

```python
# run params
SECTION = 'vae'
RUN_ID = '0001'
DATA_NAME = 'digits'
RUN_FOLDER = 'run/{}/'.format(SECTION)
RUN_FOLDER += '_'.join([RUN_ID, DATA_NAME])

if not os.path.exists(RUN_FOLDER):
    os.mkdir(RUN_FOLDER)
    os.mkdir(os.path.join(RUN_FOLDER, 'viz'))
    os.mkdir(os.path.join(RUN_FOLDER, 'images'))
    os.mkdir(os.path.join(RUN_FOLDER, 'weights'))

MODE =  'build' #'load' #
```

```python
AE = Autoencoder(
    input_dim = (28,28,1),
    encoder_conv_filters = [32,64,64,64],
    encoder_conv_kernel_size = [3,3,3,3],
    encoder_conv_strides = [1,2,2,1],
    decoder_conv_t_filters = [64,64,32,1],
    decoder_conv_t_kernel_size = [3,3,3,3],
    decoder_conv_t_strides = [1,2,2,1],
    z_dim = 2)

#AE.save(RUN_FOLDER)
```

```python
AE.encoder.summary()
```

```python
AE.decoder.summary()
```

```python
LR = 0.0005
BATCH_SIZE = 32
INIT_EPOCH = 0
```

```python
AE.compile(LR)
```

```python
AE.train(
    x_train[:1000],
    batch_size=BATCH_SIZE,
    epochs=400,
    run_folder=RUN_FOLDER,
    initial_epoch=INIT_EPOCH
)
```

```python
from utils.loaders import load_model
import numpy as np
```

```python
RAE = load_model(Autoencoder, RUN_FOLDER)
```

```python
n_to_show = 10
example_idx = np.random.choice(range(len(x_test)), n_to_show)
example_images = x_test[example_idx]
```

```python
z_points = AE.encoder.predict(example_images)
```

```python
z_points
```

```python
recons_images = AE.decoder.predict(z_points)
```

```python
import matplotlib.pyplot as plt 
```

```python
fig = plt.figure(figsize = (15,3))

fig.subplots_adjust(hspace = 0.4, wspace = 0.4)

for i in range(n_to_show):
    img = example_images[i].squeeze()
    ax = fig.add_subplot(2, n_to_show, i+1)
    ax.axis('off')
    ax.text(0.5, -0.35, str(np.round(z_points[i],1)), fontsize = 10, ha = 'center',
           transform=ax.transAxes)
    ax.imshow(img, cmap = 'gray_r')

for i in range(n_to_show):
    img = recons_images[i].squeeze()
    ax = fig.add_subplot(2, n_to_show, i+n_to_show+1)
    ax.axis('off')
    ax.imshow(img, cmap='gray_r')
```
