```python
n_test_samples = 500
n_train_samples = 1000
n_val_samples = 500
```

```python
import os, shutil
cur_dir_path = os.getcwd()
or_dataset_path = os.path.join(os.path.join(cur_dir_path, "dataset"),"train")
reduced_dataset_path = os.path.join(cur_dir_path,"dataset_reduced")

#paths

train_path = os.path.join(reduced_dataset_path,"train")
train_cats_path = os.path.join(train_path,"cats")
train_dogs_path = os.path.join(train_path,"dogs")

test_path = os.path.join(reduced_dataset_path,"test")
test_cats_path = os.path.join(test_path,"cats")
test_dogs_path = os.path.join(test_path,"dogs")

val_path = os.path.join(reduced_dataset_path,"validation")
val_cats_path = os.path.join(val_path,"cats")
val_dogs_path = os.path.join(val_path,"dogs")
```

```python
def check_n_create_dir(path):
    if not os.path.exists(path):
        os.mkdir(path)

#get current path
check_n_create_dir(reduced_dataset_path)

#train path
check_n_create_dir(train_path)

check_n_create_dir(train_cats_path)

check_n_create_dir(train_dogs_path)

#test path
check_n_create_dir(test_path)

check_n_create_dir(test_cats_path)

check_n_create_dir(test_dogs_path)

#validation path
check_n_create_dir(val_path)

check_n_create_dir(val_cats_path)

check_n_create_dir(val_dogs_path)

#copying train images
index = 0
fnames = ["cat.{}.jpg".format(i) for i in range(index,index + n_train_samples)]

for fname in fnames:
    src = os.path.join(or_dataset_path, fname)
    dst = os.path.join(train_cats_path, fname)
    shutil.copyfile(src,dst)
    
fnames = ["dog.{}.jpg".format(i) for i in range(index,index + n_train_samples)]

for fname in fnames:
    src = os.path.join(or_dataset_path, fname)
    dst = os.path.join(train_dogs_path, fname)
    shutil.copyfile(src,dst)

#copying validation images
index += n_train_samples

fnames = ["cat.{}.jpg".format(i) for i in range(index,index + n_val_samples)]

for fname in fnames:
    src = os.path.join(or_dataset_path, fname)
    dst = os.path.join(val_cats_path, fname)
    shutil.copyfile(src,dst)
    
fnames = ["dog.{}.jpg".format(i) for i in range(index,index + n_val_samples)]

for fname in fnames:
    src = os.path.join(or_dataset_path, fname)
    dst = os.path.join(val_dogs_path, fname)
    shutil.copyfile(src,dst)

#copying test images
index += n_val_samples

fnames = ["cat.{}.jpg".format(i) for i in range(index,index + n_test_samples)]

for fname in fnames:
    src = os.path.join(or_dataset_path, fname)
    dst = os.path.join(test_cats_path, fname)
    shutil.copyfile(src,dst)
    
fnames = ["dog.{}.jpg".format(i) for i in range(index,index + n_test_samples)]

for fname in fnames:
    src = os.path.join(or_dataset_path, fname)
    dst = os.path.join(test_dogs_path, fname)
    shutil.copyfile(src,dst)
```

```python
print("training cat images : ",len(os.listdir(train_cats_path)))
print("training dog images : ",len(os.listdir(train_dogs_path)))

print("validation cat images : ",len(os.listdir(val_cats_path)))
print("validation dog images : ",len(os.listdir(val_dogs_path)))

print("test cat images : ",len(os.listdir(test_cats_path)))
print("test dog images : ",len(os.listdir(test_dogs_path)))
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(150,150),
    batch_size=20,
    class_mode="binary")

val_generator = val_datagen.flow_from_directory(
    val_path,
    target_size=(150,150),
    batch_size=20,
    class_mode="binary")
```

```python
from model import model_v1

model = model_v1()

model.fit_generator(
    train_generator,
    steps_per_epoch=100,
    epochs = 300,
    validation_data=val_generator,
    validation_steps=50)

model.save("cats_and_dogs.h5")
```
