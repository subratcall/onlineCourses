```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline 
```

```python
df = pd.read_csv('WMT.csv', index_col='Date', parse_dates = True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
df['Open'].plot(figsize=(16,6))
df.rolling(90).mean()['Open'].plot()
```

```python
df['Close 30 Day MA'] = df['Close'].rolling(window=90).mean()
df[['Close 30 Day MA','Close']].plot(figsize=(16,6))
```

```python
df['Close'].expanding().mean().plot(figsize=(16,6))
```

BollingerBands

```python
#Close 20 MA 
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()

#Upper Band = 20 MA + 2* STD(20) 
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())

#Lower Band = 20 MA - 2 * STD(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())
```

```python
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].plot(figsize=(16,6), xlim=)
```
