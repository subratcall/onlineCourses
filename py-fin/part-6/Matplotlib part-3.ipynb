```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()

x = np.linspace(0,5,11)
y = x**2

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,y,
        color='purple', 
        linewidth = 3, 
        alpha = 0.9, 
        linestyle='--',
        marker = 'o',
        markersize=20, 
        markerfacecolor = 'yellow',
        markeredgecolor = 'green'
       )
```
