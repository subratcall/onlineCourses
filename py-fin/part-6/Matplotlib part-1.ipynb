```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x, y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

multiplot

```python
plt.subplot(1,2,1) #num row, num col, cum col youre plotting 
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

object oriented method

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # the list takes left bottom weight height

axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])#correspond to a certain plot
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.3])

axes1.plot(x,y)
axes1.set_title('Larger One')
axes2.plot(y,x)
axes2.set_title('Smaller One')
```
