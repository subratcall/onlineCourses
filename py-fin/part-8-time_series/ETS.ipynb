```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
airline = pd.read_csv('airline_passengers.csv', index_col='Month')
```

```python
airline.plot()
```

```python
airline.dropna(inplace = True)
airline.index = pd.to_datetime(airline.index)
```

```python
from statsmodels.tsa.seasonal import seasonal_decomposepose
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'], model = 'explonential')
```
